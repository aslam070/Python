### 2. Question 2 ( 5 marks)
You are given a dataset that contains information about hours studied and corresponding test scores. Your task is to perform linear and polynomial regression analysis to determine which model best fits the data. Complete the following steps:

1. Plotting the Data:

Create a scatter plot of the data to visualize the relationship between hours studied and test scores.
2. Fit and Evaluate a Linear Model:

Fit a linear regression model to the data.
Calculate and plot the residuals to check for any patterns.
Compute the R-squared and Mean Squared Error (MSE) for the linear model.
3. Fit and Evaluate a Polynomial Model:

Fit a polynomial regression model of degree 2 to the data.
Compute the R-squared and MSE for the polynomial model.
4. Calculate Adjusted R-squared:

Calculate the Adjusted R-squared for both the linear and polynomial models.
5. Compare Models:

Compare the Adjusted R-squared values to determine which model fits the data better.
6. Predict

Predict the test score for a student who studied for 9 hours using both models.

Instructions:
Fill in the blanks (_____) to complete the code.
Carefully read each instruction and write the correct code or values in the blank spaces.
After completing the test, run the code and verify the output.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression      # Fill in the correct model import for linear regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error  # Fill in the correct metric import for Mean Squared Error

# Data: Hours studied and corresponding test scores
x = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([50, 55, 60, 68, 80, 85, 89, 92])

# Part 1: Plotting the data points to check for linearity
plt.scatter(x, y, color='blue')    # Use an appropriate plotting function for scatter plot
plt.title("Hours Studied vs Test Score")
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")
plt.show()

# Fit a linear model
linear_model = LinearRegression()  # Initialize the linear regression model
linear_model.fit(x, y)  # Train the linear model using independent and dependent data
y_pred_linear = linear_model.predict(x)

# Plotting residuals to check for patterns
residuals = y -y_pred_linear# Calculate the residuals
plt.scatter(x, residuals, color='red')
plt.axhline(y=0, color='blue', linestyle='--')
plt.title("Residual Plot")
plt.xlabel("Hours Studied")
plt.ylabel("Residuals")
plt.show()

# Calculate R-squared for the linear model
r2_linear = r2_score(y, y_pred_linear)  # Calculate R-squared using actual and predicted values
mse_linear = mean_squared_error(y, y_pred_linear)    # Calculate MSE for the linear model
print(f"Linear Model R^2: {r2_linear:.4f}, MSE: {mse_linear:.4f}")

# Fit a polynomial model (degree 2)
poly = PolynomialFeatures(degree=2)
x_poly =poly.fit_transform(x)   # Transform the x data to polynomial features

poly_model = LinearRegression()  # Initialize the linear regression model again for polynomial regression
poly_model.fit(x_poly, y)  # Train the polynomial model
y_pred_poly = poly_model.predict(x_poly)   # Predict using the polynomial model

# Calculate R-squared for the polynomial model
r2_poly = r2_score(y, y_pred_poly)  # Calculate R-squared for the polynomial model
mse_poly = mean_squared_error(y, y_pred_poly)    # Calculate MSE for the polynomial model
print(f"Polynomial Model R^2: {r2_poly:.4f}, MSE: {mse_poly:.4f}")

# Calculate Adjusted R-squared
n = len(x)  # Number of samples
p_linear = 1  # Linear model has 1 predictor (x)
p_poly = 2    # Polynomial model (degree 2) has 2 predictors (x and x^2)

# Adjusted R-squared for Linear Model
adjusted_r2_linear = 1 - (1 -r2_linear ) * (n - 1) / (n - p_linear - 1)

# Adjusted R-squared for Polynomial Model
adjusted_r2_poly = 1 - (1 - r2_poly) * (n - 1) / (n - p_poly - 1)

# Print Adjusted R-squared values
print(f"Adjusted R^2 (Linear Model): {adjusted_r2_linear:.4f}")
print(f"Adjusted R^2 (Polynomial Model): {adjusted_r2_poly:.4f}")

# Compare and choose the best model
if adjusted_r2_poly > adjusted_r2_linear:
    print("Polynomial model fits better based on Adjusted R^2.")
else:
    print("Linear model fits better based on Adjusted R^2.")

# Bonus: Predict score for 9 hours of study
hours_studied = np.array([[9]])
predicted_linear = linear_model.predict(hours_studied)
predicted_poly = poly_model.predict(poly.fit_transform(hours_studied))

print(f"Predicted score with linear model for 9 hours: {predicted_linear[0]:.2f}")
print(f"Predicted score with polynomial model for 9 hours: {predicted_poly[0]:.2f}")
```

You are given a dataset containing monthly temperature readings for the year 2023. Your tasks are to perform the following analyses:

1. Plot the Temperature Data:

Create a line plot of the temperature data over the year.
2. Calculate and Plot Rolling Mean:

Compute the rolling mean of the temperature with a window size of 3 months.
3. Plot both the original temperature data and the rolling mean on the same graph.
4. Check for Stationarity:

Use the Augmented Dickey-Fuller test to check if the temperature time series is stationary.
5. Print the ADF statistic, p-value, and critical values.
6. Interpret the p-value to determine if the time series is stationary.
Complete the following code by filling in the appropriate functions and parameters:

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller  # Import the Augmented Dickey-Fuller test function

# Sample time series data: Monthly temperature for 2023
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Temperature': [30, 32, 35, 40, 45, 50, 55, 60, 55, 50, 45, 35]
})
data.set_index('Date', inplace=True)

# Part 1: Plot the Temperature Data
plt.plot(data.index, data['Temperature'], marker='o', color='blue')
plt.title("Monthly Temperature Data for 2023")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.show()

# Part 2: Calculate and Plot Rolling Mean
rolling_mean = data['Temperature'].rolling(window=3).mean()  # Calculate rolling mean with a window of 3 months

plt.plot(data.index, data['Temperature'], marker='o', color='blue', label='Temperature')
plt.plot(data.index, rolling_mean, color='red', label='Rolling Mean')
plt.title("Temperature and Rolling Mean")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.legend()
plt.show()

# Part 3: Check for Stationarity
result =adfuller(data['Temperature'])  # Perform the Augmented Dickey-Fuller test for stationarity
print("ADF Statistic:", result[0])  # Print the test statistic
print("p-value: ",result[1])  # Print the p-value
print("Critical Values:",result[4])  # Print the critical values
# print(result)

if result[1] <= 0.05:  # Check if the p-value is less than or equal to 0.05
    print("The time series is stationary ")  # The time series is stationary
else:
    print("The time series is not stationary ")  # The time series is not stationary
```

### Question 4 ( 3 marks )
You are given a dataset containing information about customer income and whether they purchased a product. Your task is to perform logistic regression to analyze how income affects the likelihood of purchase. Complete the following steps:

Data Preparation:

Load the dataset and prepare it by defining the feature and target variable.
Model Training:

Train a logistic regression model using the Income feature to predict whether the customer purchased the product.
Model Evaluation:

Evaluate the model's performance on a test set and report the accuracy score.
Interpretation:

Print the coefficient for the Income feature and the intercept of the model, both formatted to four decimal places.
Fill in the missing parts in the code below to achieve the tasks described:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression  # Import the logistic regression model
from sklearn.metrics import accuracy_score  # Import the accuracy score metric

# Sample data: Income and Purchase (1 = purchased, 0 = not purchased)
data = pd.DataFrame({
    'Income': [30000, 32000, 50000, 60000, 55000, 35000, 40000, 65000],
    'Purchased': [0, 0, 1, 1, 1, 0, 0, 1]
})

# Part 1: Data Preparation
X = data[['Income']]  # Define feature (Income)
y = data['Purchased']  # Define target variable (Purchased)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# Use train_test_split to split the data into training and testing sets (test_size determines the proportion of the test set)

# Part 2: Train a Logistic Regression Model
model = LogisticRegression()  # Initialize the logistic regression model
model.fit(X_train, y_train)  # Train the model using the training data

# Part 3: Evaluate the Model
y_pred = model.predict(X_test)  # Predict the target variable for the test set
accuracy = accuracy_score(y_test, y_pred)  # Calculate the accuracy score comparing the true labels to the predicted labels
print(f'Accuracy: {accuracy:.4f}')  # Print the accuracy score rounded to 4 decimal places

# Part 4: Interpret the Coefficients
coefficients = model.coef_  # Retrieve the coefficients of the logistic regression model
intercept = model.intercept_  # Retrieve the intercept of the model

# Print coefficient for Income with 4 decimal places
print(f'Coefficient for Income: {coefficients[0][0]:.4f}')

# Print intercept with 4 decimal places
print(f'Intercept: {intercept[0]:.4f}')
```
